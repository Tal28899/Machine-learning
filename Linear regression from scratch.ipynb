```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1,2,3,4,5])
y = np.array([3,4,2,5,6])
```

```python
# intializing weight and bias 
w = 0 # aka slope
b = 0 # aka intercept
```

```python
# setting hyperparameters
learning_rate = 0.01 # small steps to move towards global minima
epochs = 1000 # use in gradient decendent
n = len(X)# use in cost function
```

```python
# Gradient decendent
for i in range(epochs):
    y_pred = w * X + b # equation of a line
    error = y - y_pred # error value
    dw = (-2/n) * sum (X * error)# derivative of cost funtion with respect to w
    db = (-2/n) * sum ( error)# derivative of cost funtion with repect to b
    w = w - learning_rate * dw# updating the value of w
    b = b - learning_rate * db# updating the value of b
```

```python
#best_fit line
print(f'Weight = {w}\n Bias = {b}')  
```
