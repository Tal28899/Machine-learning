```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_diabetes
```

```python
# loading dataset
diabetes = load_diabetes()
print(diabetes)
X = diabetes.data[:,np.newaxis ,2]# selecting only one feature and changing it into 2D array(aka column vector)
y =diabetes.target# class label
```

```python
X.shape
```

```python
y.shape
```

```python
# splitting data for training and testing
X_train,X_test,y_train,y_test =train_test_split (X , y , test_size = 0.2 , random_state = 23)
```

```python
#model selection
model = LinearRegression ()
model.fit(X_train,y_train)
```

```python
# model prediction (Global minima) 
y_pred = model.predict(X_test)
```

```python
# measuring accuracy
mse = mean_squared_error(y_test,y_pred)
```

```python
print( 'Mean_Squared_Error: : ' , mse)
print('Weight (w) :',model.coef_[0])# weights are stored in array
print('Bias (b) : ',model.intercept_)
```

```python
plt.scatter(X_test,y_test,color = 'blue',label = "Actual Test Data")
plt.plot(X_test,y_pred , color = 'red',linewidth  = 2 ,label = 'Predictedline')
plt.xlabel('Bmi Feature')
plt.ylabel('Disease Progression')
plt.title("LinearRegression on diabetes")
plt.legend()
plt.show()          

```
